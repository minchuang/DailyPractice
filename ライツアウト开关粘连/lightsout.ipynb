```python
def switch(cur_lights, pos, N, M):
    new_lights = list(cur_lights)
    row = pos // M
    left = row * M
    right = (row + 1) * M - 1
    # center
    new_lights[pos] ^= 1
    # left
    if left < pos <= right:
        new_lights[pos - 1] ^= 1
    # right
    if left <= pos < right:
        new_lights[pos + 1] ^= 1
    # top
    if row > 0:
        new_lights[pos - M] ^= 1
    # bottom
    if row < N - 1:
        new_lights[pos + M] ^= 1
    return new_lights


def search(cur_pos, cur_lights, N, M, hist=[]):
    result = []
    for pos in range(cur_pos, N * M):
        new_hist = hist + [pos]
        new_lights = switch(cur_lights, pos, N, M)
        if sum(new_lights) == 0:
            result.append(new_hist)
        else:
            result += search(pos + 1, new_lights, N, M, new_hist)
    return result
```
